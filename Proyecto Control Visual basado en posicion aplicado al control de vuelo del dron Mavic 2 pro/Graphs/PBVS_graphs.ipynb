```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def plot_graphs(time, pos, vel):
    plt.figure()
    plt.plot(time, pos[:,0], label = 'x')
    plt.plot(time, pos[:,1], label = 'y')
    plt.plot(time, pos[:,2], label = 'z')
    plt.title('Posiciones en los tres ejes con respecto al  tiempo')
    plt.xlabel('Tiempo')
    plt.ylabel('Metros')
    plt.legend()
    plt.imshow

    plt.figure()
    plt.plot(time, vel[:,0], label = '$\dot{x}$')
    plt.plot(time, vel[:,1], label = '$\dot{y}$')
    plt.plot(time, vel[:,2], label = '$\dot{z}$')
    plt.title('Velocidades en los tres ejes con respecto al  tiempo')
    plt.xlabel('Tiempo')
    plt.ylabel('Metros/segundo') 
    plt.legend()
    plt.imshow

    plt.figure()
    plt.plot(time, pos[:,3], label = '$\psi$')
    plt.title('Valor de la rotacion yaw')
    plt.xlabel('Tiempo')
    plt.ylabel('Radiandes') 
    plt.legend()
    plt.imshow

    plt.figure()
    plt.plot(time, vel[:,3], label = '$\dot{\psi}$')
    plt.title('Valor de la rotacion yaw')
    plt.xlabel('Tiempo')
    plt.ylabel('Radiandes/segundo') 
    plt.legend()
    plt.imshow
```

## Simulation 1

```python
time = np.loadtxt("time_1.txt", dtype = float)
pos = np.loadtxt("pos_1.txt", dtype = float)
vel = np.loadtxt("vel_1.txt", dtype = float)

plot_graphs(time ,pos, vel)
```
